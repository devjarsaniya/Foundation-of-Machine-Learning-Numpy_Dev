```python
ar=np.array([10,-5,12,-4,-3,29])
result=np.where(ar>0,ar,0)#(condition,true,false)
print(result)
```

```python
arr=np.array([[78,85,90],[45,60,55],[88,92,95],[30,40,35],[65,70,75]])
print(arr)
```

```python
print(arr.shape)
print(arr.size)
print(arr.dtype)
```

```python

marks_student_2 = arr[1]
print(marks_student_2)
```

```python

marks_subject_3 = arr[:, 2]
print(marks_subject_3)
```

```python

mark_student_1_subject_2 = arr[0, 1]
print(mark_student_1_subject_2)
```

```python

average_marks_per_student = np.mean(arr, axis=1)
print(average_marks_per_student)


average_marks_per_subject = np.mean(arr, axis=0)
print(average_marks_per_subject)
```

```python

updated_arr = arr + 5

print(arr)

print(updated_arr)
```

```python

status_matrix = np.where(arr >= 50, 1, 0)

print(arr)

print(status_matrix)
```

```python

standardized_grades_arr = np.where(arr < 40, 0, arr)

print(arr)

print(standardized_grades_arr)
```

```python
grade = np.where(average_marks_per_student < 50, 0, np.where(average_marks_per_student < 70, 6, np.where(average_marks_per_student < 85 , 8, 10)))
print(grade)
```
