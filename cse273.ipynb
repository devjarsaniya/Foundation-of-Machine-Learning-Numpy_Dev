DATAFRAME in Pandas

```python
data={
    "Name":["Aman","krish","Karan"],
    "Age":[22,23,21],
    "Marks":[85,90,88]

}


print(data)
type(data)
marks=pd.DataFrame(data)
print(marks)
```

```python
marks=pd.DataFrame(
    data,
    index=["stu1","stu2","stu3"]
)

print(marks)
```

```python
marks.set_index("Name",inplace=True) #if we are setting inplace= true it means we are changing the original object itself.
print(marks)
```
