Create DataFrame (20 Records)

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Spotify_Hrs":[1,2,3,4,5,2,3,4,5,3,2,3,4,5,2,3,4,5,3,2],
    "YT_Music":[1,1,2,2,3,1,2,3,2,1,2,3,3,1,2,2,3,2,1,2],
    "Podcast_Hrs":[0,1,1,2,2,1,1,2,2,1,0,1,2,2,1,1,2,2,1,1],
    "Work_Hours":[6,7,8,7,6,7,8,7,6,7,6,7,8,7,6,7,8,7,6,7],
    "Productivity":[6,7,8,7,6,7,8,7,6,7,6,7,8,7,6,7,8,7,6,7]
}

df = pd.DataFrame(data)
# Renaming 'Spotify_Hrs' to 'Spotify_Hours' to match later usage and resolve KeyError
df.rename(columns={"Spotify_Hrs": "Spotify_Hours"}, inplace=True)

print(df)
```

Statistical Analysis

```python
print("Mean:\n",df.mean())

print("\nMedian:\n",df.median())

print("\nStandard Deviation:\n",df.std())

print("\nCorrelation:\n",df.corr())
```

Scatter Plot (Music vs Productivity)

```python
plt.scatter(df["YT_Music"], df["Productivity"])

plt.xlabel("Music Hours")
plt.ylabel("Productivity")
plt.title("Music vs Productivity")

plt.show()
```

Histogram (Podcast Hours)

```python
plt.hist(df["Podcast_Hrs"], bins=5)

plt.xlabel("Podcast Hours")
plt.ylabel("Frequency")
plt.title("Distribution of Podcast Hours")

plt.show()
```

Bar Chart (Spotify Hours)

```python
plt.figure()

plt.bar(range(1,21), df["Spotify_Hours"])

plt.xlabel("Student ID")
plt.ylabel("Spotify Hours")

plt.title("Spotify Usage by Students")

plt.show()
```
