Assume DataFrame df already exists with columns:

Name

Department

Role

Experience

Salary

```python
import pandas as pd
import numpy as np


data = {
    "Name": ["Dev", None, "Meet", "Dhruv", None,"Yash", "Krish", None, "Parth", "Nisarg"],

    "Department": ["IT", "HR", None, "IT", "Finance",None, "IT", "Sales", None, "IT"],

    "Role": ["Developer", "Analyst", None, "Developer", "Analyst","Manager", None, "Analyst", "Manager", None],

    "Experience": [2, None, 8, 3, None,10, 4, None, 9, 1],

    "Salary": [60000, None, 90000, 65000, 70000,None, 62000, 72000, None, 58000],

    "Score": [85, None, 90, 88, 92,95, None, 96, 91, None]
}

df = pd.DataFrame(data)

print(df)
```

```python
print(df.groupby('Department')['Salary'].mean())
```

```python
print(df.groupby('Role')['Score'].mean())
```

```python
print(df.groupby('Department')['Score'].count())
```

```python
print(df.groupby('Department').size())
```

```python
print(df.groupby('Department')['Salary'].agg(['mean', 'max', 'min', 'count']))
```

```python
print(df.groupby('Role')['Score'].agg(['mean', 'max', 'min', 'count']))
```

```python
print(df.groupby(['Department', 'Role'])['Salary'].mean())
```

```python
print(df.groupby(['Department', 'Experience'])['Score'].mean())
```
