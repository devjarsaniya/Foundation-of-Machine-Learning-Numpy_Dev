```python
#axis in numpy

arr=np.array([[10,12,14],[4,5,6]])
print(arr)
```

```python
#sum
print(np.sum(arr))
```

```python
#column wise sum
print(np.sum(arr,axis=0))
```

```python
#row wise sum
print(np.sum(arr,axis=1))
```

```python
ar=np.array([10,-5,12,-4,-3,29])
result=np.where(ar>0,ar,0)#(condition,true,false)
print(result)
```

```python
arr=np.array([1,2,3,4,5,6])
result=np.where(arr%2!=0,-1,arr)#(condition,true,false)
print(result)
```

```python
marks=np.array([10,20,32,54,76])
result=np.where(marks>35,"pass","fail")#(condition,true,false)
print(result)
```

```python
data=np.array([10,20,30,40,50])

mean_value=np.mean(data)
result=np.where(data>=mean_value,-1,data)#(condition,true,false)
print(result)
```

```python
mat=np.array([[10,60,30],[80,20,90]])

result=np.where(mat>=50,"pass","fail")#(condition,true,false)
print(result)
```
