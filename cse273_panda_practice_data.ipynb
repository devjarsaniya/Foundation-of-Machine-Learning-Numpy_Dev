```python
import pandas as pd
```

```python
df=pd.DataFrame({
    "name":["Aman","Saurav","Karan","Shubham","Abhishek",None,"Sameer"],
    "age":[22,23,None,22,25,24,26],
    "marks": [85,None,88,99,56,72,None],
    "subject":["maths","physics","maths","chemistry","maths","physics","maths"]
})
print(df)
```

```python
df = pd.DataFrame(df)

print(df)
```

```python
print(df.shape)
```

```python
print(df.dtypes)
```

```python
print(df.describe())
```

```python
view = df[["name", "marks"]]
print(view)
```

```python
filtered = df[df['marks'] > 80]
print(filtered)
```

```python
complex_filtered = df[(df['age'] > 22) & (df['marks'] > 80)]
print(complex_filtered)
```

```python
count_subject = df.groupby('subject').size()
print(count_subject)
```

```python
sorted_df = df.sort_values(by='marks', ascending=False)
print(sorted_df)
```

```python
top2 = df.nlargest(2, 'marks')
print(top2)
```

```python
lowest = df.nsmallest(1, 'marks')
print(lowest)
```

```python
print(df.isna().sum())
```

```python
missing_marks = df[df['marks'].isna()]
print(missing_marks)
```

```python
cleaned_df = df.dropna()
print(cleaned_df)
```

```python
print("Original Shape:", df.shape)
print("Cleaned Shape:", cleaned_df.shape)
```
